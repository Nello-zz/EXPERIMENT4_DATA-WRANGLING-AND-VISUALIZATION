#### PART 1: Creating and Manipulating DataFrames

```python
import pandas as pd
```

```python
# Load the required excel file into a DataFrame
board_df = pd.DataFrame(pd.read_excel('board2.xlsx'))
board_df
```

<span style="color:green; font-size:16px; font-weight:bold">A. Create a DataFrame Named Instru</span>

**Expected Output:**
- The DataFrame should contain the following columns: Name, GEAS, and Electronics.
- Only include rows where the Electronics score is greater than 70.
- Filter the data to include only students whose Track is Instrumentation and whose Hometown is Luzon.

```python
# Extracting from DataFrame board_df
Instru = board_df[(board_df['Track'] == 'Instrumentation') & (board_df['Hometown'] == 'Luzon') & (board_df['Electronics']>70)]
# Simplify the output to display only the 'Name', 'GEAS', and 'Electronics' columns
Instru = Instru[['Name', 'GEAS', 'Electronics']]
Instru
```

```python
# Adding a new column 'Average' by calculating the row-wise mean of 'Math', 'Electronics', 'GEAS', and 'Communication' scores
board_df['Average'] = board_df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
board_df
```
